```python
res_first = [0 for i in range(25)]
print(len(res_first))
for i in range(500):
    a = res[i][0]
    if a <= 8:
        res_first[0] += 1
        continue
    if a > 32:
        res_first[24] += 1
        print(a)
        continue
    for j in range(19):
        if 8+j < a <= 9+j:
            res_first[j+1] += 1
            continue
print(res_first)
```
